```python
import pandas as pd
```

```python
sv = pd.read_excel('../input/selected_variables.xlsx')
sv
```

```python
for year in range(2009, 2022):
    variables = pd.read_csv(f'../variables/variables_acs5_{year}.csv')
    for sv_index, sv_row in sv.iterrows():
        label_2022 = sv_row['2022_label']
        for var_index, var_row in variables.iterrows():
            if var_row['Label'] == label_2022 or var_row['Label'].replace('Number!!', '') == label_2022:
                if sv_row['2022_name'].endswith('PE'):
                    sv.at[sv_index, f'{year}_name'] = var_row['Name'][:-1] + 'PE'
                else:
                    sv.at[sv_index, f'{year}_name'] = var_row['Name']
                break
    
```
