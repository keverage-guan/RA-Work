```python
import zipfile
import pandas as pd
import os
import re
```

```python
# dict for columns
data = {'filename': [], 'place': [], 'year': []}
```

```python
# for each file in new_elections
for zip in os.listdir('new_elections/dataverse'):
    zip_ref= zipfile.ZipFile('new_elections/dataverse/' + zip)

    files = zip_ref.namelist()

    # for each file in the zip
    for file in files:
        if file == 'Minneapolis 2013-board of estimation and taxation cvr.csv':
            city = 'Minneapolis'
            year = '2013'
        else:
            # remove .csv and tokenize
            tokens = file.replace('.csv', '').split('_')
            # add spaces before capital letters in city
            city = re.sub(r"\B([A-Z])", r" \1", tokens[0])
            year = tokens[1][-4:]
        # add to dict
        data['filename'].append(file)
        data['place'].append(city)
        data['year'].append(year)

    zip_ref.close()
```

```python
# convert data to df
df = pd.DataFrame(data)
df
```
