```python
import pandas as pd
```

```python
df = pd.read_csv('../output/final_data.csv')
df.head()
```

```python
# SANITY CHECKS
for index, row in df.iterrows():
# check that Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Bachelor's degree or higher is nan or is < Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over
    if row['Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Bachelor\'s degree or higher'] > row['Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over']:
        print(index, row['Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over!!Bachelor\'s degree or higher'], row['Estimate!!EDUCATIONAL ATTAINMENT!!Population 25 years and over'], 'check 1')
# check that Percent!!EMPLOYMENT STATUS!!Civilian labor force!!Unemployment Rate is between 0 and 100
    if row['Percent!!EMPLOYMENT STATUS!!Civilian labor force!!Unemployment Rate'] < 0 or row['Percent!!EMPLOYMENT STATUS!!Civilian labor force!!Unemployment Rate'] > 100:
        print(index, 'check 2')
# check that Estimate!!EMPLOYMENT STATUS!!Population 16 years and over!!In labor force!!Armed Forces < Estimate!!RACE!!Total population
    if row['Estimate!!EMPLOYMENT STATUS!!Population 16 years and over!!In labor force!!Armed Forces'] >= row['Estimate!!RACE!!Total population']:
        print(index, 'check 3')
    if row['Estimate!!FERTILITY!!Number of women 15 to 50 years old who had a birth in the past 12 months'] >= row['Estimate!!RACE!!Total population']:
        print(index, 'check 4')
# check that Estimate!!INCOME AND BENEFITS!!Total households!!Mean household income (dollars) is between 10000 and 200000
    if row['Estimate!!INCOME AND BENEFITS!!Total households!!Mean household income (dollars)'] < 10000 or row['Estimate!!INCOME AND BENEFITS!!Total households!!Mean household income (dollars)'] > 300000:
        print(index, row['Estimate!!INCOME AND BENEFITS!!Total households!!Mean household income (dollars)'], 'check 5')
# same for Estimate!!INCOME AND BENEFITS!!Total households!!Median household income (dollars)
    if row['Estimate!!INCOME AND BENEFITS!!Total households!!Median household income (dollars)'] < 10000 or row['Estimate!!INCOME AND BENEFITS!!Total households!!Median household income (dollars)'] > 200000:
        print(index, 'check 6')
# check that Estimate!!INDUSTRY!!Civilian employed population 16 years and over!!Agriculture, forestry, fishing and hunting, and mining < Estimate!!INDUSTRY!!Civilian employed population 16 years and over
    if row['Estimate!!INDUSTRY!!Civilian employed population 16 years and over!!Agriculture, forestry, fishing and hunting, and mining'] >= row['Estimate!!INDUSTRY!!Civilian employed population 16 years and over']:
        print(index, 'check 7')
# check that Estimate!!U.S. CITIZENSHIP STATUS!!Foreign-born population!!Not a U.S. citizen < Estimate!!U.S. CITIZENSHIP STATUS!!Foreign-born population
    if row['Estimate!!U.S. CITIZENSHIP STATUS!!Foreign-born population!!Not a U.S. citizen'] >= row['Estimate!!U.S. CITIZENSHIP STATUS!!Foreign-born population']:
        print(index, 'check 8')
```
