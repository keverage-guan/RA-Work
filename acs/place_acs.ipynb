```python
# Import modules
import pandas as pd
import cleaning
import utils
```

```python
# open elections.xlsx
elections = pd.read_excel('elections.xlsx')

# open elections_with_geocodes.xlsx
geocodes = pd.read_excel('elections_with_geocodes.xlsx')

groups = ['DP02', 'DP03', 'DP05']
key = 'ad8851f3bf6aaf76923ec4119b6f714cdfaa87d9'
```

```python
acs_data = []
```

```python
for index, row in elections.iterrows():
    # row['geo_filename'] is either a string or nan, continue if not nan
    if not pd.isnull(row['geo_filename']):
        continue

    if index % 10 == 0:
        print(f"{index} of {len(elections)}")

    year = row['year']

    # get row where filename is equal to row['filename']
    geocode_row = geocodes[geocodes['filename'] == row['filename']]
    for_param = utils.get_place_for_param(geocode_row)
    data_dict = utils.query(year, groups, for_param, key)
    data_dict['filename'] = row['filename']

    acs_data.append(data_dict)
```

```python
acs_data = cleaning.dicts_to_df(acs_data)
# order so that geocode and year are first
acs_data = acs_data[['filename'] + [col for col in acs_data.columns if col not in ['filename']]]
```

```python
# replace all (X), -, or N with nan
cleaning.replace_error_values(acs_data)
cleaning.combine_percent_columns(acs_data)
cleaning.remove_invalid_columns(acs_data)
cleaning.remove_nonaggregatables(acs_data)
acs_data
```

```python
# save acs_data as place_acs_data.csv
acs_data.to_csv('place_acs_data.csv', index=False)
```
