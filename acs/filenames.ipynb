```python
import pandas as pd
import geopandas as gpd
import os
```

```python
# import key _rows.xlsx
key_rows = pd.read_excel(r'D:\Users\User\Downloads\RA Work\acs\key_rows.xlsx')

key_rows
```

```python
# add columns to key_rows named geo_filename and tract_filename
key_rows['geo_filename'] = ''
key_rows['tract_filename'] = ''
```

```python
# iterate through each row in key_rows
for index, row in key_rows.iterrows():
    city = row['city'].lower().replace(" ", "_")
    geography = row['geography'].replace(" ", "_")
    election_year = row['year']

    if geography != 'city':
        path = fr"D:\Users\User\Downloads\RA Work\acs\{city}"
        files = os.listdir(path)

        geo_year = 1990
        tract_year = 1900
        for file in files:
            # if file ends with 4 numbers and .csv
            if file[-8:-4].isdigit() and file[-4:] == '.csv':
                year = int(file.split("_")[-1].split(".")[0])
                if 'tract' not in file:
                    if year > geo_year and year <= election_year:
                        geo_year = year
                else:
                    if year > tract_year and year <= election_year:
                        tract_year = year

        geo_file = f"{city}_{geography}_{geo_year}.csv"
        tract_file = f"{city}_tracts_{tract_year}.csv"

        key_rows.at[index, 'geo_filename'] = geo_file
        key_rows.at[index, 'tract_filename'] = tract_file
```
